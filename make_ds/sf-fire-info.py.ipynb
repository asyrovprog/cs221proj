```python
import pandas as pd
```

```python
ds = pd.read_csv("../ds/sf-fire-counts-train.csv", low_memory=False)
ds.head()
```

```python
ds.info()
```

```python
%matplotlib inline
# only in a Jupyter notebook
import matplotlib.pyplot as plt
ds.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
%matplotlib inline
# only in a Jupyter notebook
import matplotlib.pyplot as plt

y = ds['count']
x = [i for i in range(0, len(y))]

plt.plot(x, y)
plt.show()
    
```

```python
%matplotlib inline
# only in a Jupyter notebook
import matplotlib.pyplot as plt

for i in range(0, 10):
    y = ds['count'][i * 200: i * 200 + 168]
    x = [i for i in range(0, len(y))]
    
    plt.plot(x, y)
    plt.show()
```
